```python
# Configuration du WebDriver
options = webdriver.ChromeOptions()
options.headless = True
driver = webdriver.Chrome(options=options)

# Accéder à la page
url = "https://fr.shein.com/Women-Clothing-sc-017172961.html?..."
driver.get(url)
time.sleep(15)

# Trouver tous les éléments de produit
product_elements = driver.find_elements(By.CSS_SELECTOR, ".product-card")

# Créer des listes pour stocker les données
product_names = []
product_prices = []
product_promotions = []
product_sales = []

# Itérer sur chaque élément de produit
for product in product_elements:
    # Extraire le nom du produit
    name_element = product.find_element(By.CSS_SELECTOR, "a.goods-title-link--jump")
    name = name_element.get_attribute('aria-label') if name_element else "Nom inconnu"
    product_names.append(name)

    # Extraire le prix du produit
    price_elements = product.find_elements(By.CSS_SELECTOR, "span.normal-price-ctn__sale-price")
    price = price_elements[0].text if price_elements else "Prix inconnu"
    product_prices.append(price)

    # Vérifier et extraire les informations de promotion, si présentes
    promotion_elements = product.find_elements(By.CSS_SELECTOR, "div.product-card__discount-label")
    promotion = promotion_elements[0].text if promotion_elements else "Pas de promotion"
    product_promotions.append(promotion)

    # Vérifier et extraire les informations de vente
    sales_elements = product.find_elements(By.CSS_SELECTOR, "p.product-card__selling-proposition-text")
    if sales_elements:
        sales_text = sales_elements[0].text.replace('k', '000')
        match = re.search(r'\d+', sales_text)
        sales_number = int(match.group()) if match else 0
    else:
        sales_number = 0  # Si aucun élément de vente n'est trouvé
    product_sales.append(sales_number)

# Fermer le WebDriver
driver.quit()

# Afficher les résultats
print(product_names)
print(product_prices)
print(product_promotions)
print(product_sales)
print(len(product_names))
print(len(product_prices))
print(len(product_promotions))
print(len(product_sales))
```

```python
    # Création du DataFrame
df = pd.DataFrame({
    'Nom': product_names,
    'Prix': product_prices,
    'Nombre de ventes récemment':product_sales,
    'Réduction': product_promotions
                    })

filename = "Produit_Féminin.csv"
df.to_csv(filename, index=False, encoding='utf-8')

print(f"Le fichier {filename} a été sauvegardé avec succès.")
```
